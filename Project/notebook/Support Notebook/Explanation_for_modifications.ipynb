At the very beginning, we were going to match all the kindle version products with the corresponding book products with the help of **asin** because we had assumed that each asin correspond to one book and the same asin in kindle product dataset and book dataset could present the description of the two versions respectively. However, we found that all the data rows with the same asin owned the same description in both datasets, which means that if there exists a kindle version for a book, the description of this book in book product dataset is also the description of kindle version. Thus, we can not use asin to match books.

Here is an example of the same asin in the two description datasets.

```python
df_kindle_meta.loc[df_kindle_meta.index == '1603420304']
```

```python
df_book_meta.loc[df_book_meta.index == '1603420304']
```

As one of our main purpose is to compare the difference among the reviews of the two versions, we still would like to believe that the reviews of the same asin can be distinct in the two review datasets. Nevertheless, although the number of reviews for the same asin is different from the two datasets (number of book reviews is greater than that of kindle product reviews), those reviews which are not in the kindle product review dataset are still reviews concerning the kindle version after checking Amazon website. So, we totally abandon the idea of matching books.

Here is an example of the reviews of the same asin in two review datasets. For asin 'B000FA64PK', there are 8 reviews in kindle product review dataset (all of them are in book review dataset as well) and 9 reviews in book review dataset. After checking Amazon website, the one only in book review dataset is still concerning kindle version.

```python
df_kindle_review.loc[df_kindle_review['asin'] == 'B000FA64PK']
```

```python
df_book_review.loc[df_book_review['asin'] == 'B000FA64PK']
```

As we can not match the two versions of a book, we try to use category to carry out our project. We try to classify all the kindle books (from kindle product description dataset) and those books which doesn't have kindle version (from book description dataset) respectively by category. And then, we try to build the connection between kindle products and books by category. 

```python
kindle_cat = []
for idx in df_kindle_meta.index:
    cat = df_kindle_meta.loc[idx].categories
    for catcat in cat:
        if catcat[0]=='Kindle Store':
            kindle_cat.append([idx]+catcat[1:])
        else:
            kindle_cat.append([idx]+catcat)
```

```python
df_kindle_category = pd.DataFrame(kindle_cat)
df_kindle_category.columns = ['asin','1st','2nd','3rd','4th','5th','6th','7th']
df_kindle_category.set_index('asin',inplace=True)
df_kindle_category.head()
```

Tha DataFrame above shows us the classification result for kindle product according to the different category levels. However, after discarding all the kindle product data from book descriotion dataset, we unfortunately find almost all the books don't possess precise category for classification. 

```python
df_book_meta_without_kindle = pd.read_pickle('pickleData/book_metadata_without_kindle.pkl')
df_book_meta_without_kindle.head()
```

Hence, we should say we can not build any connection between kindle product dataset and book dataset.
